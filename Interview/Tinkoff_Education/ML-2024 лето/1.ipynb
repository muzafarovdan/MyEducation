```python
n = int(input())
list_marks = list(map(int, input().split()))
```

```python
max_fives = -1

for i in range(n - 6):
    segment = list_marks[i:i+7]  # Выбираем семь оценок начиная с i-ой позиции
    if 2 not in segment and 3 not in segment:  # Проверяем, что в отрезке нет 2 и 3
        fives_count = segment.count(5)  # Считаем количество пятерок в отрезке
        max_fives = max(max_fives, fives_count)  # Обновляем максимальное количество пятерок

if max_fives == -1:  # Если не нашли ни одного подходящего отрезка
    print(-1)
else:
    print(max_fives)
```
