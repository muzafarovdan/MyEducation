```python
n = int(input())
list_marks = list(map(int, input().split()))
```

```python
max_fives = -1

for i in range(n - 6):
    segment = list_marks[i:i+7]  #      i- 
    if 2 not in segment and 3 not in segment:  # ,     2  3
        fives_count = segment.count(5)  #     
        max_fives = max(max_fives, fives_count)  #    

if max_fives == -1:  #       
    print(-1)
else:
    print(max_fives)
```
