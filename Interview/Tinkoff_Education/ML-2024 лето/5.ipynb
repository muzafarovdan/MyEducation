```python
def dfs(field, row, col, mushrooms):
    if row < 0 or row >= len(field) or col < 0 or col >= len(field[0]) or field[row][col] == 'W':
        return mushrooms
    
    # Если текущая клетка содержит гриб, увеличиваем количество грибов
    if field[row][col] == 'C':
        mushrooms += 1
    
    # Массив смещений для возможных ходов
    moves = [(1, -1), (1, 0), (1, 1)]
    
    max_mushrooms = mushrooms
    # Проверяем все возможные ходы
    for dr, dc in moves:
        new_row, new_col = row + dr, col + dc
        if 0 <= new_row < len(field) and 0 <= new_col < len(field[0]):
            max_mushrooms = max(max_mushrooms, dfs(field, new_row, new_col, mushrooms))
    
    return max_mushrooms

# Чтение входных данных
n = int(input())
field = [input() for _ in range(n)]

# Поиск пути с максимальным количеством клеток с грибами
max_mushrooms = 0
for j in range(len(field[0])):
    if field[0][j] != 'W':
        max_mushrooms = max(max_mushrooms, dfs(field, 0, j, 0))

print(max_mushrooms)
```
