```python
def print_directories(directory, depth=0):
    for key in sorted(directory.keys()):
        print("  " * depth + key)
        if directory[key]:
            print_directories(directory[key], depth + 1)

n = int(input())
directories = {}

for _ in range(n):
    path = input().split('/')
    current = directories
    for folder in path:
        if folder not in current:
            current[folder] = {}
        current = current[folder]

print_directories(directories)
```
