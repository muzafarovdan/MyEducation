```python
n, q = map(int, input().split())
array = list(map(int, input().split()))

for _ in range(q):
    a = input().split()
    if a[0] == '?':
        l = int(a[1])-1
        r = int(a[2])-1
        k = int(a[3])
        b = int(a[4])
        array_now = array[l:r+1]
        maximum = max(array_now)
        i = array.index(maximum)
        minimum = min(maximum, k * (i+1) + b)
        print(minimum)
    elif a[0] == '+':
        l = int(a[1])-1
        r = int(a[2])-1
        x = int(a[3])
        for i in range(l, r+1):
            array[i] += x
        
```
