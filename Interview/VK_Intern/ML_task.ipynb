```python
df_train = df_train.drop(columns='Unnamed: 0')
df_train.head(5)
```

```python
X = df_train.drop(columns='im')
y = df_train['im']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7, stratify=y)
```

```python
X_train
```

## Слишком много фичей, надо отбирать

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)

```

```python
X_train_scaler = pd.DataFrame(data=X_train_scaler)
X_test_scaler = pd.DataFrame(data=X_test_scaler)
X_train_scaler
```

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
pca = PCA(svd_solver='full', n_components=66)
pca.fit(X_train_scaler)
```

```python
array = pca.explained_variance_ratio_
np.where(np.cumsum(array) >= 0.9)[0][0] + 1
```

```python
X_train_pca = pca.transform(X_train_scaler)
X_test_pca = pca.transform(X_test_scaler)
```

```python
X_train_res = pd.DataFrame(data=X_train_pca)
X_test_res = pd.DataFrame(data=X_test_pca)
X_train_res
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Предполагая, что X это ваша таблица данных, загрузите ее в DataFrame pandas

# Пример:
# X = pd.read_csv('ваш_файл.csv')

# Построение матрицы корреляций
correlation_matrix = X_train_res.corr()

# Построение тепловой карты корреляций
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

## Модель Log Reg

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.metrics import classification_report


model_LR = LogisticRegression().fit(X_train_res, y_train)
pred = model_LR.predict(X_test_res)
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)

print(classification_report(y_test, pred, digits=4))
```

## Модель Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model_RF = RandomForestClassifier(max_depth=5,
                                  random_state=0, 
                                  n_estimators=8,
                                  min_samples_split=11,
                                  min_samples_leaf=4)
model_RF.fit(X_train_res, y_train)
pred = model_RF.predict(X_test_res)
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)

print(classification_report(y_test, pred, digits=4))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from tqdm import tqdm

# Определите сетку гиперпараметров для GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Создайте объект GridSearchCV
grid_search = GridSearchCV(RandomForestClassifier(random_state=0), param_grid, cv=5)

# Выполните поиск по сетке с помощью tqdm для отображения прогресса
with tqdm(total=len(param_grid['n_estimators']) * len(param_grid['max_depth']) * len(param_grid['min_samples_split']) * len(param_grid['min_samples_leaf'])) as pbar:
    grid_search.fit(X_train_res, y_train)
    pbar.update(1)

# Получите лучшие гиперпараметры
best_params = grid_search.best_params_

# Получите лучшую модель
best_model = grid_search.best_estimator_

# Сделайте прогнозы с лучшей моделью
pred = best_model.predict(X_test_res)

# Оцените модель
print("Best Parameters:", best_params)
print("\nClassification Report:")
print(classification_report(y_test, pred, digits=4))
```
