```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
df_train = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/fish_train.csv')
df = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/fish_reserved.csv')
df_train.head(5)
```

```python
weight = df_train.pop('Weight')
species = df_train.pop('Species')
height = df_train.pop('Height')
width = df_train.pop('Width')
df_train.head(5)
```

```python
pca = PCA(n_components=1, svd_solver='full')
data_transformed = pca.fit_transform(df_train) # перевод в базис главных компонент
lengths = data_transformed
lengths
```

```python
df_train = df_train.drop(['Length1', 'Length2', 'Length3'], axis=1)
```

```python
df_train['Lengths'] = lengths
df_train['Species'] = species
df_train['Height'] = height
df_train['Width'] = width
df_train.head(5)
```

```python
df_train = pd.get_dummies (df_train, columns=['Species'], drop_first=True)
df_train.head(5)
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True)
poly = poly.fit_transform(df_train)
```

```python
df_transgender = pd.DataFrame(data=poly)
df_transgender.head(5)
```

```python
regressor = LinearRegression()
model = regressor.fit(df_transgender, weight)
# Y_pred = regressor.predict(test_reserv)
```
