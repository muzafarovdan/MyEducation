```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/pulsar_stars_new_ex.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df1 = df[(df['TG'] == 0)]
df1 = df1[df1['MIP'] >= 58.2890625]
df1 = df1[df1['MIP'] <= 170.1328125]
df1
```

```python
df2 = df[(df['TG'] == 1)]
df2 = df2[df2['MIP'] >= 8.2734375]
df2 = df2[df2['MIP'] <= 127.5703125]
df2
```

```python
df = pd.concat([df1, df2])
df
```

```python
df['MIP'].mean()
```

```python
df = df.sort_values(by=['SIP'])
df
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='TG')
y = df['TG']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
X_train
```

```python
X_train['EKIP'].mean()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
X_train_scaler.T[4].mean()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
model_LR = LogisticRegression(random_state = 7)
model_LR.fit(X_train_scaler,y_train)
pred = model_LR.predict(X_test_scaler)
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
f1_score(pred, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN = KNeighborsClassifier(n_neighbors=3)
model_KNN.fit(X_train_scaler,y_train)
pred = model_KNN.predict(X_test_scaler)
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
f1_score(pred, y_test)
```

```python
df_train = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/MNIST_train.csv')
df_test = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/MNIST_test.csv')
```

```python
df_train = df_train[(df_train['label'] == 0) | (df_train['label'] == 2)]
df_test = df_test[(df_test['label'] == 0) | (df_test['label'] == 2)]
```

```python
df_train.head(5)
```

```python
X_train = df_train.drop(columns='label')
y_train = df_train['label']

X_test = df_test.drop(columns='label')
y_test = df_test['label']
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(svd_solver='full')
pca.fit(X_train)
```

```python
var = pca.explained_variance_ratio_

np.where(np.cumsum(var) >= 0.83)[0][0] + 1
```

```python
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
X_train[12][0]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_GBC = GradientBoostingClassifier(n_estimators=500, learning_rate=0.8,max_depth=2, random_state=25).fit(X_train, y_train)
```

```python
pred = model_GBC.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred, digits=4))
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
estimators = [('rf', RandomForestClassifier(random_state=25)),('svc', SVC(random_state=25))]
final_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression(random_state=25))
```

```python
final_model.fit(X_train, y_train)
pred = final_model.predict(X_test)

print(classification_report(y_test, pred, digits=4))
tn, fp, fn, tp = confusion_matrix(pred, y_test).ravel()
print(tn, fp, fn, tp)
```
