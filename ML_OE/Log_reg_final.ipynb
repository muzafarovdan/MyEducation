```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
data = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/titanic_train.csv')
data_test = pd.read_csv('/Users/muzafarov/Desktop/MyEducation/ML_OE/titanic_reserved.csv', delimiter=',')
```

```python
data.head(10)
```

```python
data_test.head(10)
```

```python
data['fam_size'] = data['sibsp'] + data['parch']
data = data.drop(columns=['sibsp', 'parch'])

data_test['fam_size'] = data_test['sibsp'] + data_test['parch']
data_test = data_test.drop(columns=['sibsp', 'parch'])
```

```python
data = data.drop(columns=['ticket','home.dest','cabin'])
data_test = data_test.drop(columns=['ticket','home.dest','cabin'])
```

```python
data.head(10)
```

```python
data_test.head(10)
```

```python
list_name = []
for i in data['name']:
    one = i.index(',')
    two = i.index('.')
    list_name.append(i[one+2:two])

data['honorific'] = list_name

data.replace({'honorific': ['Rev','Col', 'Dr', 'Major', 'Don', 'Capt']}, 'Mr', inplace=True)
data.replace({'honorific': ['Dona', 'Countess', 'the Countess']}, 'Mrs', inplace=True)
data.replace({'honorific': ['Mlle', 'Ms']}, 'Miss', inplace=True)
new_table = data['honorific'].value_counts()

data[data.honorific == 'Mr'] = data[data.honorific == 'Mr'].fillna(data[data.honorific == 'Mr'].age.mean())
data[data.honorific == 'Miss'] = data[data.honorific == 'Miss'].fillna(data[data.honorific == 'Miss'].age.mean())
data[data.honorific == 'Mrs'] = data[data.honorific == 'Mrs'].fillna(data[data.honorific == 'Mrs'].age.mean())
data[data.honorific == 'Master'] = data[data.honorific == 'Master'].fillna(data[data.honorific == 'Master'].age.mean())

data.head(10)
```

```python
list_name = []
for i in data_test['name']:
    one = i.index(',')
    two = i.index('.')
    list_name.append(i[one+2:two])

data_test['honorific'] = list_name

data_test.replace({'honorific': ['Rev','Col', 'Dr', 'Major', 'Don', 'Capt', 'Sir', 'Mme', 'Jonkheer']}, 'Mr', inplace=True)
data_test.replace({'honorific': ['Dona', 'Countess', 'the Countess', 'Lady']}, 'Mrs', inplace=True)
data_test.replace({'honorific': ['Mlle', 'Ms']}, 'Miss', inplace=True)
new_table = data_test['honorific'].value_counts()

data_test[data_test.honorific == 'Mr'] = data_test[data_test.honorific == 'Mr'].fillna(data_test[data_test.honorific == 'Mr'].age.mean())
data_test[data_test.honorific == 'Miss'] = data_test[data_test.honorific == 'Miss'].fillna(data_test[data_test.honorific == 'Miss'].age.mean())
data_test[data_test.honorific == 'Mrs'] = data_test[data_test.honorific == 'Mrs'].fillna(data_test[data_test.honorific == 'Mrs'].age.mean())
data_test[data_test.honorific == 'Master'] = data_test[data_test.honorific == 'Master'].fillna(data_test[data_test.honorific == 'Master'].age.mean())

data_test.head(10)
```

```python
data = data.drop(columns='name')
data_test = data_test.drop(columns='name')
data_test.head(10)
```

```python
data = pd.get_dummies (data, drop_first=True)
data_test = pd.get_dummies (data_test, drop_first=True)
```

```python
data_test = data_test.drop(columns=['embarked_37.21739130434783', 'embarked_C'])
data_test.head(10)
```

```python
model = LogisticRegression(random_state=1)
model.fit(X_train,y_train)
pred = model.predict(X_test)
```
